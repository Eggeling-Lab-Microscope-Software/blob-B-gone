```python
gen = np.random.Generator(np.random.PCG64(42))

fig,ax = plt.subplots(1,1,figsize=(5, 5), dpi = 300)
mean = [0,0]
cov = np.array([[1,0],[0,1]])*1e-16
x,y = gen.multivariate_normal(mean, cov, 10000).T
ax.hist2d(x,y, bins = 50, cmap = 'plasma')
ax.set_title('Blob Simulation')
ax.set_xlabel('x [m]')
r = 10e-9
theta = np.linspace(0, 2*np.pi, 100)
ax.plot(r*np.cos(theta), r*np.sin(theta), '--', c='white', linewidth = 1, label = 'r: 10 nm')
ax.plot(3*r*np.cos(theta), 3*r*np.sin(theta), '1', c='white', linewidth = 0.5, label = 'r: 30 nm')
ax.set_ylabel('y [m]')
ax.axis('equal')
ax.set_facecolor('darkblue')
ax.legend(loc = 'upper right', fontsize = 8, facecolor = 'black', edgecolor = 'white', labelcolor = 'white')

plt.show()
```
